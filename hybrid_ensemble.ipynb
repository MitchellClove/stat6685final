```python
import pandas as pd

cat_df = pd.read_csv("test_probabilities/catboost_probs.csv")
cnn_df = pd.read_csv("test_probabilities/cnn_probability.csv")
lstm_df = pd.read_csv("test_probabilities/lstm_probability.csv")

cnn_df = cnn_df.rename(columns={"probability": "cnn_prob"})
lstm_df = lstm_df.rename(columns={"probability": "lstm_prob"})

df = cat_df.merge(cnn_df, on="object_id")
df = df.merge(lstm_df, on="object_id")

w_cat = 0.65
w_cnn = 0.20
w_lstm = 0.15

df["ensemble_prob"] = (
    w_cat * df["catboost_prob"]
    + w_cnn * df["cnn_prob"]
    + w_lstm * df["lstm_prob"]
)

df["prediction"] = (df["ensemble_prob"] >= 0.5).astype(int)

submission = df[["object_id", "prediction"]]
submission.to_csv("hybrid_submission_65_20_15_NEW.csv", index=False)

print("Submission saved.")
print(submission.shape)
print(df.head())
# Count number of predicted TDEs
num_tde = submission["prediction"].sum()

print("Number of predicted TDEs:", num_tde)
```
